```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
Dataset = pd.read_csv('Position_Salaries.csv')
x=Dataset.iloc[:,1:-1].values
y=Dataset.iloc[:,-1].values
```

```python
y=y.reshape(len(y),1)
```

```python
from sklearn.preprocessing import StandardScaler
Std_x= StandardScaler()
Std_y= StandardScaler()
x=Std_x.fit_transform(x)
y=Std_y.fit_transform(y)
```

```python
from sklearn.svm import SVR
regressor= SVR(kernel='rbf')
regressor.fit(x,y)
```

```python
Std_y.inverse_transform(regressor.predict(Std_x.transform([[6.5]])))
```

```python
plt.scatter(Std_x.inverse_transform(x),Std_y.inverse_transform(y),color='red')
plt.plot(Std_x.inverse_transform(x),Std_y.inverse_transform(regressor.predict(x)),color='blue')
plt.show()
```
